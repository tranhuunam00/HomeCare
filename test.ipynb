```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo ảnh 64x64 RGB (giá trị từ 0–255)
img_array = np.random.randint(0, 256, (1, 3, 1), dtype=np.uint8)

# Hiển thị ảnh
plt.imshow(img_array)
plt.title("Generated 9x9 Image")
plt.axis('off')
plt.show()

# In thông tin
print("Shape:", img_array.shape)
print("Data type:", img_array.dtype)

# In góc trên trái 5x5
print("Top-left 5x5 pixels:")
print(img_array[:5, :5])
```
